```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = []
val = []
with open('../exps/try/log') as f:
    for line in f.readlines():
        line = line.strip()
        if False:
            pass
        if line.startswith("| Train Done"):
            fields = line.split(',')
            record = []
            for field in fields[1:]:
                k, v = field.strip().split(':')
                record.append(float(v))
            train.append(record)
        elif line.startswith("* Val Done"):
            fields = line.split(',')
            record = []
            for field in fields[1:]:
                k, v = field.strip().split(':')
                record.append(float(v))
            val.append(record)
train = np.array(train)
val = np.array(val)
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)


x_val = np.arange(len(val)) + 1
x_train = np.arange(len(train)) + 1
ax1.plot(x_train, train[:, 0], label='hitAll_train')
ax1.plot(x_val, val[:, 1], label='hitAll_val')
ax1.plot(x_train, train[:, 1], label='hitOne_train')
ax1.plot(x_val, val[:, 2], label='hitOne_val')


ax2.plot(x_train, train[:, -1], label='loss_train')
ax2.plot(x_val, val[:, -2], label='loss_val')

ax1.legend(loc='best', fontsize=20)
ax2.legend(loc='best', fontsize=20)
plt.show()
```
