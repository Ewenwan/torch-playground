```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
mnist = [
    ['Float32', 0.95],
    ['16Bits', 0.95],
    ['8Bits', 0.95],
    ['6Bits', 1.07],
    ['4Bits', 2.88],
]
vgg19 = [
    ['Float32', 28.74, 10.01],
    ['16Bits', 28.75, 10.01],
    ['8Bits', 29.40, 10.27],
    ['6Bits', 79.06, 59.42]
]
x_mnist, y_mnist = zip(*mnist)
x_vgg19, y_vgg19_top1, y_vgg19_top5 = zip(*vgg19)
index_mnist = np.arange(len(mnist)) + 0.4
index_vgg19 = np.arange(len(vgg19)) + 0.4
```

```python
fig = plt.figure(figsize=(14, 7))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

rects1 = ax1.bar(index_mnist, y_mnist, 0.4, alpha=0.4)
ax1.set_xlabel('Bitwidth', fontsize=15)
ax1.set_ylabel('Error (%)', fontsize=15)
ax1.set_title('Error Rate vs. Bitwidth of LeNet', fontsize=15)
ax1.set_xticks(index_mnist+0.2)
ax1.set_xticklabels(x_mnist, fontsize=15)
ax1.set_ylim(0, 3.5)
for i, rect in enumerate(rects1):
    height = rect.get_height()
    ax1.text(rect.get_x() + rect.get_width()/2.0, 1.02*height, "{0:.2f}".format(mnist[i][1]), ha='center', va='bottom', fontsize=12)
    
rects2 = ax2.bar(index_vgg19, y_vgg19_top1, 0.4, alpha=0.4, label='Top1 Error', color='b')
rects3= ax2.bar(index_vgg19 + 0.4, y_vgg19_top5, 0.4, alpha=0.4, label='Top5 Error', color='r')
ax2.set_xlabel('Bitwidth', fontsize=15)
#ax2.set_ylabel('Error (%)', fontsize=15)
ax2.set_title('Error Rate vs. Bitwidth of VGG19', fontsize=15)
ax2.set_xticks(index_vgg19+0.4)
ax2.set_xticklabels(x_vgg19, fontsize=15)
ax2.set_ylim(0, 90)
ax2.legend(loc='upper left')
for i, rect in enumerate(rects2):
    height = rect.get_height()
    ax2.text(rect.get_x() + rect.get_width()/2.0, 1.02*height, "{0:.2f}".format(vgg19[i][1]), ha='center', va='bottom', fontsize=12)
for i, rect in enumerate(rects3):
    height = rect.get_height()
    ax2.text(rect.get_x() + rect.get_width()/2.0, 1.02*height, "{0:.2f}".format(vgg19[i][2]), ha='center', va='bottom', fontsize=12)
    
    
fig.tight_layout()
plt.savefig('../vis/lenet.jpg')
plt.show()
```
